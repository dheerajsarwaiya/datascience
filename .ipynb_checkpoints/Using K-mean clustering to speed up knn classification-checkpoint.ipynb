CLustering is used to find relationship and meaning in data. In most of the cases, semi supervised learning is recommended. Use clustering to define classes and then use supervised learning to predict.

Will experiment with k-mean here. Will divide the data into K clusters. It generate K centroids, then calculate distance of each featureset to the centroids. First centroid be assiened as 0, second as 1 and so on. Once one round is done, it will calculate new mean of the groups and set new centroids. Then repeat

```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
from sklearn.cluster import KMeans
```

```python
style.use('ggplot')
```

```python
X = np.array([[1, 2],
              [1.5, 1.8],
              [5, 8],
              [8, 8],
              [1, 0.6],
              [9, 11]])


plt.scatter(X[:, 0],X[:, 1], s=150, linewidths = 5, zorder = 10)
plt.show()
```

```python
clf = KMeans(n_clusters=2)
clf.fit(X)
```

```python
centroids = clf.cluster_centers_
labels = clf.labels_
```

```python
colors = ["g.","r.","c.","y."]
for i in range(len(X)):
    plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize = 10)
plt.scatter(centroids[:, 0],centroids[:, 1], marker = "x", s=150, linewidths = 5, zorder = 10)
plt.show()
```

Focus on the survival column. My program will decide on the groups. I am doing flat clustering, basically dividing into two groups. Later let machine decide on the groups.

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import pandas as pd
```

I have to install xlrd package here

```python
df = pd.read_excel('titanic.xls')
print(df.head())
```

handle non numerical data.. Drop name.. 


```python
df.drop(['body','name'], 1, inplace=True)
```

```python
df.convert_objects(convert_numeric=True)
```

```python
df.fillna(0, inplace=True)
```

```python
def handle_non_numerical_data(df):
    columns = df.columns.values
    for column in columns:
        text_digit_vals = {}
        def convert_to_int(val):
            return text_digit_vals[val]

        if df[column].dtype != np.int64 and df[column].dtype != np.float64:
            column_contents = df[column].values.tolist()
            unique_elements = set(column_contents)
            x = 0
            for unique in unique_elements:
                if unique not in text_digit_vals:
                    text_digit_vals[unique] = x
                    x+=1

            df[column] = list(map(convert_to_int, df[column]))

    return df
```

```python
df = handle_non_numerical_data(df)
print(df.head())
```

```python
X = np.array(df.drop(['survived'], 1).astype(float))
y = np.array(df['survived'])
```

```python
clf = KMeans(n_clusters=2)
clf.fit(X)
```

```python
correct = 0
for i in range(len(X)):
    predict_me = np.array(X[i].astype(float))
    predict_me = predict_me.reshape(-1, len(predict_me))
    prediction = clf.predict(predict_me)
    if prediction[0] == y[i]:
        correct += 1

print(correct/len(X))
```
